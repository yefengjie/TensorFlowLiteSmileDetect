### 0.2、接下来我们处理一下随机数
由于模型训练存在随机性,每次训练的结果不一定一致,所以我们需要提前设置一个随机种子,保证每次模型训练的结果都是一致的.随机种子不用刻意的去选取,随便选择一个数就好,一个好的模型不会因为随机种子的不同而表现出过大的差异

```python
import os
import random
import tensorflow as tf
import numpy as np
# 随便设置一个随机种子
random_seed= 10
# 禁止hash随机
os.environ['PYTHONHASHSEED']=str(random_seed)
np.random.seed(random_seed)
random.seed(random_seed)
tf.random.set_seed(random_seed)
```

## 1、数据采集及预处理
这里我们采取 [开源数据集](https://github.com/hromi/SMILEsmileD), 数据概览: 64*64灰度图. 共9475个笑脸图片,3690个没有笑的图片

注意:在实际项目中,我们需要人工采集数据,并进行标记

### 1.1、图片尺寸缩放以及灰度化
为了后面训练,我们将图片尺寸缩放为:IMG_SIZE*IMG_SIZE,并转化为灰度图片. 

如果遇到图片不是正方形的情况,可以将图片按照短边裁剪为正方形,也可以按照长边扩充为正方形,空白部分填充像素0

```python
import cv2
import imutils
from imutils import paths
from keras.preprocessing.image import img_to_array
from PIL import Image

# 图片参与训练尺寸
IMG_SIZE=32

# 训练集路径
dataset_dir = os.path.abspath("SMILEs")
print("预览正常图片")
display(Image.open('normal.jpg'))
print("预览笑脸图片")
display(Image.open('smile.jpg'))
data=[]
labels = []
for imagePath in sorted(list(paths.list_images(dataset_dir))):
    image = cv2.imread(imagePath)
    # 转为灰度图像
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # 将图片尺寸缩小至IMG_SIZE
    image = imutils.resize(image, width=IMG_SIZE, height=IMG_SIZE)
    # 使用Keras的img_to_array转换成浮点型和（IMG_SIZE*IMG_SIZE*1），便于接下来神经网络学习
    image = img_to_array(image)
    # 添加到数据集
    data.append(image)
    # 读取图片文件夹名字
    label = imagePath.split(os.path.sep)[-2]
    # 如果是positives文件夹,标记为smile,否则标记为normal
    label = "smile" if label == "positives7" else "normal"
    labels.append(label)
print("样本维度:",np.shape(data))
print("笑脸图片数量:",labels.count("smile"),"非笑脸图片数量:",labels.count("normal"))
```

### 1.2、one hot编码
经过上面的处理

1、我们的图片数据已经变为了n * IMG_SIZE * IMG_SIZE * 1的矩阵,矩阵的每个点都是[0,255]之间的一个数字,为了后面模型训练,这个数据我们需要将他们处理为[0,1]之间的数字,简单点的做法,就是直接除以255.

2、我们的标签数据变为了normal、smile的标签,这个是不能直接参与训练的,我们需要进行one-hot编码,将他转为0和1(2个类别)

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils

print("原始像素预览:",data[0][0][0],data[0][0][1],data[0][0][2])
print("原始标签预览:",labels[0],labels[10000])
data=np.array(data,dtype="float")/255.0
labels=np.array(labels)
# 对标签进行one-hot编码
# LabelEncoder可以将标签分配一个0—类别数量之间的编码
le = LabelEncoder().fit(labels)
# transform用来标准化,将我们的标签转换为0,1的数字.2为类别数量(我们只有微笑和正常两个类别)
labels = np_utils.to_categorical(le.transform(labels), 2)
print("处理后的像素数据预览",data[0][0][0],data[0][0][1],data[0][0][2],type(data))
print("处理后的标签预览",labels[0],labels[10000],type(labels))
```

### 1.3、处理数据平衡性
经过上面的数据预览,我们发现笑脸(3690)和正常的图片(9475)数量相差太大,也就是存在数据不平衡.为了解决这个问题,所以我们后面训练的时候需要给数量添加权重,放大笑脸数据的权重为(9475/3690)倍

```python
classTotals = labels.sum(axis=0)
print("类别样本数量",classTotals)
classWeight = classTotals.max() / classTotals
classWeight = {i: classWeight[i] for i in range(2)}
print("类别样本权重", classWeight)
```

### 1.4、分割数据集
将数据按照 8:2 分割为训练数据集以及测试数据集, 训练数据集用来训练模型,测试数据集用来检验模型训练效果

```python
from sklearn.model_selection import train_test_split

(trainX, testX, trainY, testY) = train_test_split(data, labels, test_size=0.2, stratify=labels, random_state=random_seed)
print("训练数据预览:",np.shape(trainX),"测试数据预览:",np.shape(testX))
print("训练标签预览:",np.shape(trainY),"测试标签预览:",np.shape(testY))
```

## 2、模型构建与训练
我们采用LeNet进行训练,LeNet主要用户识别手写字符.[详细查看LeNet简介](https://www.jianshu.com/p/cd73bc979ba9)
![lenet](lenet.hdf5.jpg)
接下来我们来定义模型输入和输出:  

输入: IMG_SIZE * IMG_SIZE * 1

输出: 0或者1来标识是否为笑脸

```python
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.core import Dense
from keras.layers.core import Flatten
from keras.models import Sequential

output_shape=2

model = Sequential()
# 卷积层,输入图片input_shape*input_shape,卷积核5*5,滤波器20,激活层relu
model.add(Conv2D(input_shape=(IMG_SIZE, IMG_SIZE, 1), kernel_size=(5, 5), filters=20, activation='relu'))
# 池化层,池化核2*2,步长2,padding填充0
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding='same'))

# 卷积层,卷积核5*5,滤波器50,激活层relu
model.add(Conv2D(kernel_size=(5, 5), filters=50, activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=2, padding='same'))

model.add(Flatten())
model.add(Dense(500, activation='relu'))

model.add(Dense(2, activation='softmax'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=["accuracy"])
```

接下来开始训练模型,参数哦说明如下:  
class_weight:数据权重  
batch_size:梯度批次  
epochs:训练周期

```python
# 训练周期
epochs = 15
# 训练模型
history = model.fit(trainX, trainY, validation_data=(testX, testY),
              class_weight=classWeight, batch_size=64, epochs=epochs, verbose=1)  # verbose = 1显示进度条
```

## 3、模型效果评估
### 3.1、可视化我们的训练过程

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, epochs), history.history["loss"], label="loss")
plt.plot(np.arange(0, epochs), history.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, epochs), history.history["accuracy"], label="accuracy")
plt.plot(np.arange(0, epochs), history.history["val_accuracy"], label="val_accuracy")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Loss/Accuracy")
plt.legend()
plt.show()
```

### 3.2、测试报告
用测试数据集来检验我们的模型,报告指标说明如下:  

#### 3.2.1、support
类别总数.跟我们上面分割的训练集2633一致.其中1895个正常,738个笑脸

#### 3.2.2、precision & recall
precision:精确度,recall:召回率  
例如: 一堆图片里面有12只狗和10只猫,程序识别出了8只狗,但这8只狗里面其实只有5(真阳)个是真的狗,3个是猫(假阳).所以精确度为5/8(真阳/(真阳+假阳)),召回率为5/12(真阳/(真阳+假阴)).[Precision_and_recall](https://en.wikipedia.org/wiki/Precision_and_recall)

#### 3.2.3、f1-score
结合精度和召回率的度量是精度和召回率的调和平均值. f1=2*precision*recall/(precision+recall)

#### 3.2.4、support
类目数量

#### 3.2.5、accuracy
准确率,正确预测样本量/总样本量

#### 3.2.6、macro avg
宏平均,表示所有类别对应指标的平均值

例如precision: (0.95+0.80)/2=0.88
#### 3.2.7、weighted avg
权重平均，表示类别样本占总样本的比重与对应指标的乘积的累加和

例如precision: 0.95*1895/2633+0.80*738/2633=0.91 

```python
from sklearn.metrics import classification_report

predictions = model.predict(testX, batch_size=64)

print("标签预览:",le.classes_)
print("模型测试报告")
print(classification_report(testY.argmax(axis=1), predictions.argmax(axis=1), target_names=le.classes_))
```

### 3.4、用真实图片检测
上面是用我们测试的数据集进行检验,模型实际表现效果如何,还需要我们用真实的数据进行检验.这里我们从网上下载了2张图片,然后用我们的模型进行检验

```python
from keras.models import load_model

model = load_model(model_dir)
def predictImg(imagePath):
    image = cv2.imread(imagePath)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = imutils.resize(image, width=IMG_SIZE, height=IMG_SIZE)
    image = image.astype("float") / 255.0
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    (normal, smile) = model.predict(image)[0]
    display(Image.open(imagePath))
    print("预测结果->","笑脸" if(smile>normal) else "普通","\t普通 vs 笑脸:",round(normal,2)," - ",round(smile,2))
predictImg("测试普通.jpg")
predictImg("测试笑脸.jpg")
```

## 4、模型转换
上面的训练的keras模型无法直接部署到移动端,我们需要将模型转换为tflite格式以支持移动端.[转换文档](https://www.tensorflow.org/lite/guide/get_started#2_convert_the_model)

转换模型的同时还可以对模型进行优化,比如量化.详细请参考[官方文档](https://www.tensorflow.org/lite/performance/model_optimization).

我们还可以为模型添加元数据,元数据描述模型的一个标准,是了解模型输入输出信息的重要来源,由人类可读部分和机器可读部分构成.详细请参考[官方文档](https://www.tensorflow.org/lite/performance/model_optimization)

```python
model_lite_dir='lenet.tflite'
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
# Save the model.
with open(model_lite_dir, 'wb') as f:
    f.write(tflite_model)
```

### 4.1、接下来我们测试一下转换后的模型效果

```python
interpreter = tf.lite.Interpreter(model_path=model_lite_dir)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
print("输入:", str(input_details))
output_details = interpreter.get_output_details()
print("输出:", str(output_details))

def predictImgLite(imagePath):
    image = cv2.imread(imagePath)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = imutils.resize(image, width=IMG_SIZE, height=IMG_SIZE)
    image = image.astype("float") / 255.0
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    display(Image.open(imagePath))
    
    interpreter.set_tensor(input_details[0]['index'], image)
    interpreter.invoke()
    output_data = interpreter.get_tensor(output_details[0]['index'])
    (normal,smile)=output_data[0]
    print("预测结果->","笑脸" if(smile>normal) else "普通","\t普通 vs 笑脸:",round(normal,2)," - ",round(smile,2))
predictImgLite("测试普通.jpg")
predictImgLite("测试笑脸.jpg")
```

## 6、客户端数据预处理
客户端在将数据喂给模型之前也需要对输入图片进行预处理,官方提供了ImageProcessor类方便我们处理.[官方文档](https://www.tensorflow.org/lite/inference_with_metadata/lite_support)

示例代码如下:

    /**
     * 预处理图片，并转换为输入
     * @param id 原始图片资源id
     * @param processedImage 预处理图片预览组件
     */
    private fun processImgToInput(id: Int, processedImage: ImageView): ByteBuffer {
        val tmpBitmap = BitmapFactory.decodeResource(resources, id)
        // 输入的图片可能不是正方形，我们需要按照短边裁剪
        val cropSize = minOf(tmpBitmap.width, tmpBitmap.height)
        val imgProcessor = ImageProcessor.Builder()
            //按cropSize对图片进行裁剪
            .add(ResizeWithCropOrPadOp(cropSize, cropSize))
            //将裁剪好的图片缩放到输入尺寸
            .add(
                ResizeOp(
                    tfInputSize.height,
                    tfInputSize.width,
                    ResizeOp.ResizeMethod.NEAREST_NEIGHBOR
                )
            )
            //把像素值的范围进行归一化从[0,255]区间映射到[-1,1]区间
            .add(NormalizeOp(0f, 1f))
            // 转化为灰度图
            .add(TransformToGrayscaleOp())
            .build()
        val tfImageBuffer = TensorImage(DataType.FLOAT32)
        val tfImage = imgProcessor.process(tfImageBuffer.apply { load(tmpBitmap) })
        processedImage.setImageBitmap(tfImage.bitmap)
        return tfImage.buffer
    }
    
## 7、推理
经过上面的数据预处理,我们就可以对模型进行推理了.[官方文档](https://www.tensorflow.org/lite/guide/inference)

我们也可以通过切换不同的委托来看看模型的运行速度.[官方文档](https://www.tensorflow.org/lite/performance/delegates)

在下面的示例图片中,GPU delegate比NNAPI delegate运行速度稍微快一些,但实际效果每次运行和不同机型运行效果都不太一样,实际场景中我们需要多次测试选择合适委托,甚至可以动态切换委托

示例代码如下:

    /**
     * 开始推理
     * @param id 原始图片资源id
     * @param processedImage 预处理后的图片
     * @param tv 输出结果文本
     */
    private fun start(id: Int, processedImage: ImageView, tv: TextView) {
        val inputBuffer = processImgToInput(id, processedImage)
        val outputBuffer = mapOf(0 to arrayOf(FloatArray(2)))
        val timeStart = System.currentTimeMillis()
        tflite.runForMultipleInputsOutputs(arrayOf(inputBuffer), outputBuffer)
        val predictTime = System.currentTimeMillis() - timeStart
        val outputArray = (outputBuffer[0] as Array)[0]
        val result = if (outputArray[0] > outputArray[1]) "正常" else "笑脸"
        tv.text = "预测结果: $result -> 推理耗时:${predictTime}ms"
    }
